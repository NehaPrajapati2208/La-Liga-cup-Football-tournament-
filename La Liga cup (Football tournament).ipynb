# 3. Print the list of teams which came Top 5 in terms of points

```python
df['Points'] = df['Points'].astype(int)
df[['Team', 'Points']].sort_values(by = 'Points', ascending = False).head(5)
```

# 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences

```python
def Goal_diff_count():
    df['Goal_diff_count'] = df['GoalsFor'].astype(int) - df['GoalsAgainst'].astype(int)
    return df[['Team', 'Goal_diff_count']].sort_values(by = 'Goal_diff_count', ascending = False )
Goal_diff_count()
```

```python
# Maximum goal difference 
Goal_diff_count().head(1)
```

```python
# Maximum goal difference 
Goal_diff_count().tail(1)
```

# 5.Create a new column with name “Winning Percent” and append it to the data set (5 points) Percentage of Winning = (GamesWon / GamesPlayed)*100 If there are any numerical error, replace it with 0%

```python
df['Winning_Percentage'] = (df['GamesWon'].astype(int)/df['GamesPlayed'].astype(int))*100                           
```

```python
df['Winning_Percentage'].fillna(0, inplace = True)
```
